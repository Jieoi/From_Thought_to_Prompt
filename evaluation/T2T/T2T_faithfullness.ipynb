<h2> Faithfulness analysis</h2>

Helper functions:

```python
def entailment_prob(premise: str, hypothesis: str) -> float:
    """
    prob that hypothesis is entailed by premise
    """
    out = nli_pipeline(f"{premise} </s></s> {hypothesis}")
    for entry in out:
        if entry['label'] == 'ENTAILMENT':
            return entry['score']
    return 0.0
```

```python
def cosine_similarity(a: np.ndarray, b: np.ndarray) -> float:
    """
    cosine similarity between two numpy vectors symbolising the text
    """
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))
```

Calcualting the faithfulness score:

start with hallucination score

```python
def score_hallucination(originals: list[str], enhanced:  list[str], lora_whitelist: set[str]) -> pd.DataFrame:
    # place holder for records
    records = []

    # get embeddings
    orig_embs = embedder.encode(originals, convert_to_numpy=True, normalize_embeddings=True)
    enh_embs  = embedder.encode(enhanced,  convert_to_numpy=True, normalize_embeddings=True)

    # get bert score
    bert_res = bertscore_metric.compute(predictions=enhanced, references=originals, lang="en")

    # get bleu and rouge score
    for i, (orig, enh) in enumerate(zip(originals, enhanced)):
        # set to zero if either is empty, divide by 0 error
        if not orig.strip() or not enh.strip():
            bleu    = 0.0
            rouge_l = 0.0
        else:
            # lexical overlap via bleu
            bleu_res = bleu_metric.compute(predictions=[enh], references=[[orig]])
            bleu     = bleu_res["bleu"]

            # lexical longest common subseq via rouge_l
            rouge_l = rouge_metric.compute(predictions=[enh], references=[orig])["rougeL"]

        # semantic overlap
        bert_f1 = bert_res["f1"][i]

        # cos similarity
        cos_sim = cosine_similarity(orig_embs[i], enh_embs[i])  # [−1–1]

        # entailment probability
        entail  = entailment_prob(orig, enh)  # [0–1]

        # fraction of new tokens in enhanced aka novelty
        orig_toks = set(orig.split())
        enh_toks  = enh.split()
        novelty = sum(tok not in orig_toks for tok in enh_toks) / max(len(enh_toks),1)

        # fraction of <lora:…> tags not in whitelist (if error, check if lora downloaded)
        refs = re.findall(r"<lora:([\w\-]+)>", enh)
        drift = sum(r not in lora_whitelist for r in refs) / max(len(refs),1) if refs else 0.0

        records.append({
            "bleu":       bleu,
            "rouge_l":    rouge_l,
            "bert_f1":    bert_f1,
            "cosine":     (cos_sim + 1)/2, # map [−1,1] to [0,1]
            "entail":     entail,
            "novelty":    novelty,
            "lora_drift": drift
        })

    df = pd.DataFrame(records)

    # transform to “hallucination risk” and aggregate
    df["h_bleu"]    = 1 - df["bleu"]
    df["h_rouge"]   = 1 - df["rouge_l"]
    df["h_bert"]    = 1 - df["bert_f1"]
    df["h_cosine"]  = 1 - df["cosine"]
    df["h_entail"]  = 1 - df["entail"]
    df["h_novelty"] = df["novelty"]
    df["h_lora"]    = df["lora_drift"]

    # save a copy
    df["halluc_score"] = df[
        ["h_bleu","h_rouge","h_bert","h_cosine","h_entail","h_novelty","h_lora"]
    ].mean(axis=1)

    return df
```

```python
# load LoRA whitelist
wl_path = "/content/drive/MyDrive/NLP_proj/text_data/valid_loras.json"
with open(wl_path, 'r', encoding='utf-8') as f:
    raw_whitelist = json.load(f)
# extract TAG, the thing inside the lora name
lora_whitelist = {
    re.match(r"<lora:(.+)>", tag).group(1)
    for tag in raw_whitelist
    if re.match(r"<lora:(.+)>", tag)
}
```

```python
# collector to run through all folders in drive
def collect_hallucination_scores(base_path: str, model_dirs: list[str]) -> pd.DataFrame:
    records = []
    base = Path(base_path)
    orig_folder = base / "original"

    for model in model_dirs:
        if model == "original":
            continue  # skip comparing original to original
        mod_folder = base / model
        txts = sorted(
            [p for p in mod_folder.glob("*.txt") if p.stem.isdigit()],
            key=lambda p: int(p.stem)
        )
        print(f"{model}: found {len(txts)} prompts")

        for p in txts:
            idx = int(p.stem)
            orig_text = (orig_folder / f"{idx}.txt").read_text(encoding="utf-8")
            enh_text  = p.read_text(encoding="utf-8")

            # call hallucination score function
            df_one = score_hallucination(originals = [orig_text],
                                         enhanced = [enh_text],
                                         lora_whitelist = lora_whitelist)

            # add model/index by converting to dic
            row = df_one.iloc[0].to_dict()
            row.update({"model": model, "index": idx})
            records.append(row)

    # build a df with every metric and aggregate
    return (
        pd.DataFrame.from_records(records).sort_values(["model","index"], ignore_index=True)    )


# run
base_dir = "/content/drive/MyDrive/NLP_proj/text_data"
models   = ["BART_gen","deepseek_gen","QWEN_generated","t5_gen"]

df_scores = collect_hallucination_scores(base_dir, models)
print(df_scores.head())

# save
out_path = "/content/drive/MyDrive/NLP_proj/hallucination_scores_all_models.csv"
df_scores.to_csv(out_path, index=False)
print("Saved to", out_path)
```

```python
base_dir     = Path("/content/drive/MyDrive/NLP_proj/text_data")
orig_folder  = base_dir / "original"

# get originals & generations
orig_texts = []
gen_texts  = []
for _, row in df_scores.iterrows():
    idx     = int(row["index"])
    model   = row["model"]
    orig_fp = orig_folder / f"{idx}.txt"
    gen_fp  = base_dir    / model       / f"{idx}.txt"
    orig_texts.append(orig_fp.read_text(encoding="utf-8"))
    gen_texts .append(gen_fp .read_text(encoding="utf-8"))

# BERTScore precision recall
P, R, _ = bert_score(
    cands=gen_texts,
    refs =orig_texts,
    lang ="en",
    verbose=False
)
# attach to df_scores
df_scores["bert_precision"] = [p.item() for p in P]
df_scores["bert_recall"]    = [r.item() for r in R]
```

Include svo

```python
def extract_svo(doc):
    svos = []
    for token in doc:
        if token.dep_ == "nsubj" and token.head.pos_ == "VERB":
            subject = token.text.lower()
            verb = token.head.text.lower()
            for child in token.head.children:
                if child.dep_ in ("dobj", "attr", "prep"):
                    obj = child.text.lower()
                    svos.append((subject, verb, obj))
    return set(svos)
```

```python
# get svo and prescion + recall
svo_prec = []
svo_rec  = []
for orig, gen in zip(orig_texts, gen_texts):
    o_svos = extract_svo(nlp(orig))
    g_svos = extract_svo(nlp(gen))
    overlap = o_svos & g_svos
    svo_prec.append(len(overlap) / max(len(g_svos), 1))
    svo_rec .append(len(overlap) / max(len(o_svos), 1))

df_scores["svo_precision"] = svo_prec
df_scores["svo_recall"]    = svo_rec

# save
df_scores.to_csv("/content/drive/MyDrive/NLP_proj/hallucination_scores_with_all_metrics.csv", index=False)

print("New columns added:",["bert_precision","bert_recall","svo_precision","svo_recall"])
print("DataFrame now has columns:", df_scores.columns.tolist())
```
