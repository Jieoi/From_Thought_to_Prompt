```python
import pandas as pd
from datasets import Dataset

df_train = pd.read_csv("train_civitai_image_prompt_captioned_updated.csv")
df_val = pd.read_csv("val_civitai_image_prompt_captioned_updated.csv")

dataset = {
    "train": Dataset.from_pandas(df_train),
    "validation": Dataset.from_pandas(df_val)
}
```

```python
def format_prompt_response(example):
    formatted = f"<|user|>\n{example['prompt']}\n<|assistant|>\n{example['response']}"
    return {"text": formatted}

dataset["train"] = dataset["train"].map(format_prompt_response).remove_columns(
    [col for col in dataset["train"].column_names if col not in ["text"]]
)
dataset["validation"] = dataset["validation"].map(format_prompt_response).remove_columns(
    [col for col in dataset["validation"].column_names if col not in ["text"]]
)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_path = "/content/drive/MyDrive/[NLP] Project/hf_qwen1.5b-final"
tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    trust_remote_code=True,
    device_map="auto"
)
```

```python
def tokenize(example):
    return tokenizer(example["text"], padding="max_length", truncation=True, max_length=512)

tokenized_dataset = {
    "train": dataset["train"].map(tokenize, batched=True),
    "validation": dataset["validation"].map(tokenize, batched=True),
}
```

```python
from transformers import DataCollatorForLanguageModeling
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)
```

```python
!ls
```

```python
from transformers import TrainingArguments, Trainer
training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/[NLP] Project/civitai_qwen2b-promptonly",
    overwrite_output_dir=True,
    num_train_epochs=20,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=4,
    learning_rate=3e-4,
    weight_decay=0.01,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    gradient_checkpointing=True,
    greater_is_better=False,
    lr_scheduler_type="cosine",
    optim="paged_adamw_32bit",
    warmup_steps=200,
    save_total_limit=2,
    fp16=False,
    bf16=True,
    logging_steps=100,
    report_to=["none"],
)
```

```python
from transformers import EarlyStoppingCallback
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset = tokenized_dataset["validation"],
    data_collator=data_collator,
    tokenizer=tokenizer,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=4)]
)
```

```python
trainer.train()
print("Training finished.")
```

```python
model.save_pretrained("/content/drive/MyDrive/[NLP] Project/civitai_qwen2b-final_2", max_shard_size="10GB", safe_serialization=True)
tokenizer.save_pretrained("/content/drive/MyDrive/[NLP] Project/civitai_qwen2b-final_2")
```

```python
df_test = pd.read_csv("test_civitai_image_prompt_captioned_updated.csv")
test_dataset = Dataset.from_pandas(df_test)
```

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device_map="auto",
    return_full_text=False
)
```

```python
input_prompts = df_test["prompt"].tolist()
reference_responses = df_test["response"].tolist()
```

```python
formatted_inputs = [f"<|user|>\n{p}\n<|assistant|>\n" for p in input_prompts]
```

```python
from tqdm import tqdm

predictions = [
    generator(text, max_length=256)[0]['generated_text']
    for text in tqdm(formatted_inputs)
]
```

```python
import pandas as pd

df_eval = pd.DataFrame({
    "input_prompt": input_prompts,
    "reference_response": reference_responses,
    "generated_response": predictions
})

df_eval.to_csv("civitai_QWEN_test_predictions_prompt_response_2.csv", index=False)
```
